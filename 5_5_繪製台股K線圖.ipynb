```python
import yfinance as yf
import pandas as pd
import mplfinance as mpf
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')
```

繪製股價K線圖

```python
# 商品名稱
prod = "2330.TW"
# 取得全部資料
data = yf.download(prod, period="max")
# 繪製線圖
mpf.plot(data.iloc[-10:])
```

```python
# 繪製K線圖
mpf.plot(data.iloc[-10:], type="candle")
```

```python
# 查看樣式
print(mpf.available_styles())

# 調整樣式
mpf.plot(data.iloc[-10:], type="candle", style="yahoo")
```

```python
# 修改K線顏色
mcolor = mpf.make_marketcolors(up="r", down="g", inherit=True)
mstyle = mpf.make_mpf_style(base_mpf_style="yahoo", marketcolors=mcolor)
mpf.plot(data.iloc[-10:], type="candle", style=mstyle)
```
