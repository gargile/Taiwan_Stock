```python
import yfinance as yf
import pandas as pd
import matplotlib
matplotlib.rc('font', family='Microsoft JhengHei')
```

繪製股價折線圖

```python
# 商品名稱
prod = "2330.TW"
# 取得全部資料
data = yf.download(prod, period="max")
# 繪製折線圖
data['Close'].plot(label='收盤價',legend=True)
```

```python
# 繪製折線圖
(data['Close'].pct_change()+1).cumprod().plot(label='報酬率',legend=True)
(data['Adj Close'].pct_change()+1).cumprod().plot(label='還原除權息後報酬率',legend=True)
```
